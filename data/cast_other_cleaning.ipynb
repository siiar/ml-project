```python
credits_list = list()

i = 0

for movie in response_list:
    IDX = movie.get('id')  # integer
    cast = movie.get('cast') # list of dicts
    crew = movie.get('crew') # list of dicts

    count = 0
    castmembers_list = list()  # only five cast members
    for cast_member in cast:
        member = cast_member.get('name')
        castmembers_list.append(member)
        count += 1
        if count == 5:
            break  # only inner loop

    i += 1
    print(i)
    print(IDX)
    
    directors_list = list()
    crew_list = list()
    producers_list = list()
    screenplay_list = list()
    casting_list = list()
    music = list()
    for crew_member in crew:
        if crew_member.get('job') == 'Director':
            director = crew_member.get('name')
            directors_list.append(director)
        if crew_member.get('job') == 'Producer':
            producer = crew_member.get('name')
            producers_list.append(producer)
        if crew_member.get('job') == 'Screenplay':
            screenplay = crew_member.get('name')
            screenplay_list.append(screenplay)
        if crew_member.get('job') == 'Casting':
            casting = crew_member.get('name')
            casting_list.append(casting)
        if crew_member.get('job') == 'Original Music Composer':
            composer = crew_member.get('name')
            music.append(composer)


    credits_dict = {'id': IDX,
                   'cast_1': np.nan if len(castmembers_list)==0 else castmembers_list[0],
                   'cast_2': np.nan if len(castmembers_list)<=1 else castmembers_list[1],
                   'cast_3': np.nan if len(castmembers_list)<=2 else castmembers_list[2],
                   'cast_4': np.nan if len(castmembers_list)<=3 else castmembers_list[3],
                   'cast_5': np.nan if len(castmembers_list)<=4 else castmembers_list[4],
                   'director': np.nan if len(directors_list)==0 else directors_list[0],
                   'producer': np.nan if len(producers_list)==0 else producers_list[0],
                   'screenplay': np.nan if len(screenplay_list)==0 else screenplay_list[0],
                   'casting': np.nan if len(casting_list)==0 else casting_list[0],
                   'composer': np.nan if len(music)==0 else music[0]}
    credits_list.append(credits_dict)
        
```
