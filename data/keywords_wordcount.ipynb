```python
keywords = df[['id', 'overview']]
keywords.head()
```

```python
r.rank_list[:3]
```

```python
def get_top_keywords(row):
    IDX = row['id']
    overview = row['overview']
    r = Rake()
    r.extract_keywords_from_text(overview)
    mylist = r.rank_list[:3]
    keywords_list = [i[1] for i in mylist]
    print(keywords_list)
    if len(keywords_list) == 1:
        return {'id': IDX, 'keyword_1': np.nan, 'keyword_2': np.nan, 'keyword_3': np.nan}
    if len(keywords_list) == 2:
        return {'id': IDX,'keyword_1': keywords_list[0], 'keyword_2': keywords_list[1], 'keyword_3': np.nan}
    else:
        return {'id': IDX,'keyword_1': keywords_list[0], 'keyword_2': keywords_list[1], 'keyword_3': keywords_list[2]}
```

```python
some_list = list()

for i in range(len(keywords)):
    try:
        whatever = get_top_keywords(keywords.iloc[i])
        some_list.append(whatever)
    except TypeError:
        print('got some NaN value I guess....')
        empty_dict = {'id': keywords.iloc[i]['id'], 'keyword_1': np.nan, 'keyword_2': np.nan, 'keyword_3': np.nan}
        some_list.append(empty_dict)
```

```python
keywords_df = pd.DataFrame(some_list)
```

```python
keywords.iloc
```

```python
main = df.merge(keywords_df, on='id')
```

```python
main.to_csv('main_with_keywords.csv', index=False)
```
